**Identifiers** 

- iso_code - country

- year

**Outcome Variables**

- employment gap - Difference in employment rates between men and women (men % minus women %) - A value of 10 means 10% more men are employed than women. Lower = more equal.

- pay gap - Difference in gross hourly earnings between men and women as a % of men's earnings - A value of 15 means women earn 15% less per hour than men. Lower = more equal.

- inactive_care_pct - % of working-age women (20–64) who are economically inactive specifically because of caring responsibilities (children, elderly). Lower = more equal.

- women_parliament_pct — % of seats in national parliament held by women. Higher = more equal.

- female_board_share — Share of women on company boards, across all firms (public + private). Higher = more equal.

- female_senior_mgmt_share — Share of women in senior management positions across all firms. Higher = more equal.

**Policy Variables**

- childcare_spend_pct_gdp — Public spending on childcare and early education as a % of GDP. Higher = more investment in childcare infrastructure, expected to correlate with higher female employment.

- maternity_weeks — Number of weeks of paid maternity leave available to mothers. More weeks = more generous maternity policy.

- parental_paid_weeks — Number of weeks of paid parental leave available to mothers, beyond maternity leave (Interestingly, very long parental leave can hurt female employment by encouraging career exit).

- father_leave_weeks — Number of weeks of paid leave reserved exclusively for fathers. Higher = more gender-equal parental policy design. Key variable for testing whether father leave affects mother outcomes.

**World Bank WBL Policy Scores (All scored 0–100, higher = more equal laws)**

- WBL INDEX — Overall Women, Business and the Law score — average of all 8 sub-scores below. A score of 100 means full legal gender equality across all domains.

- MOBILITY — Do women have equal freedom of movement? (passport, travel, choosing where to live)

- WORKPLACE — Equal access to jobs, anti-discrimination laws, sexual harassment legislation

- PAY — Equal pay laws, right to work nights/dangerous jobs, industrial jobs

- MARRIAGE — Domestic violence laws, divorce rights, remarriage rights, obedience clauses

- PARENTHOOD — Maternity leave adequacy, paternity leave existence, dismissal protections during pregnancy

- ENTREPRENEURSHIP — Equal access to credit, contracts, business registration, bank accounts

- ASSETS — Equal property ownership, inheritance rights, asset administration during marriage

- PENSION — Equal retirement age, pension benefits accounting for childcare absences

Analytical approach - 

1. Contradiction finding — Where do high policy scores and poor outcomes coexist? Where do low policy scores and surprisingly good outcomes coexist?
2. Trajectory over position — Instead of analysing "which country scores highest?", find out "which country is improving fastest?" and "which country has plateaued despite strong policies?" 
3. Residual analysis — To run a regression predicting outcomes from policy scores. The countries with the largest residuals — those performing far better or far worse than their policy environment predicts
